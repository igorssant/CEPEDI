```python
file_name = "car_details"
car_details = read_data(file_name)
car_details.info()
```

### Remoção de espaços em branco e outras padronizações

```python
header = car_details.columns
header = fix_header_to_standarts(header)
car_details.columns = header
print(f"{car_details.columns}")
```

```python
car_details.sample(5)
```

```python
names = car_details["name"]
names = fix_header_to_standarts(names)
car_details["name"] = names
#--------------------------
owners = car_details["owner"]
owners = fix_header_to_standarts(owners)
car_details["owner"] = owners
#--------------------------
car_details.sample(5)
```

```python
car_details["name"] = car_details["name"].astype("category")
car_details["year"] = pd.to_datetime(car_details["year"]).dt.to_period("Y")
# -------------------------------------------------------
car_details["selling_price"] = car_details["selling_price"].astype("uint32")
car_details["km_driven"] = car_details["km_driven"].astype("uint32")
car_details["fuel"] = car_details["fuel"].astype("category")
# -------------------------------------------------------
car_details["seller_type"] = car_details["seller_type"].astype("category")
car_details["transmission"] = car_details["transmission"].astype("category")
car_details["owner"] = car_details["owner"].astype("category")
# -------------------------------------------------------
car_details.info()
```

```python
car_details
```

### Concat

```python
left_part = car_details[car_details.selling_price < 90001]
middle_part = car_details[car_details.km_driven > 69999]
right_part = car_details[car_details.year.dt.year > 2005]
```

```python
pd.concat([left_part, middle_part, right_part], axis = 0)
```

### Merge

```python
left_dt = car_details.copy()
to_erase_list = ["name", "selling_price", "km_driven", "owner"]

for to_erase in to_erase_list:
    left_dt.drop(to_erase, inplace = True, axis = 1)

right_dt = car_details.copy()
to_erase_list = ["fuel", "seller_type", "transmission", "owner"]

for to_erase in to_erase_list:
    right_dt.drop(to_erase, inplace = True, axis = 1)

(left_dt
    .merge(right_dt, left_on = "seller_type", right_on = "name", suffixes = ("_left", "_right"), how="outer")
)
```

### Join

```python
left_dt = car_details.copy()
to_erase_list = ["name", "selling_price", "km_driven", "owner"]

for to_erase in to_erase_list:
    left_dt.drop(to_erase, inplace = True, axis = 1)

right_dt = car_details.copy()
to_erase_list = ["fuel", "seller_type", "transmission", "owner"]

for to_erase in to_erase_list:
    right_dt.drop(to_erase, inplace = True, axis = 1)

(left_dt.set_index("year")
    .join(right_dt.set_index("name"),
    how = "outer")
)
```

### GroupBy

##### - median

```python
(car_details
    .groupby("transmission")["selling_price"]
    .median()

).astype("uint32")
```

#### - mean

```python
(car_details
    .groupby("fuel")["km_driven"]
    .mean()

).astype("float32")
```

#### - std

```python
# Carros elétricos geram np.nan, pois não há desvio padrão -> std == 0
(car_details
    .groupby("fuel")["km_driven"]
    .std()

).astype("float32").fillna(0)
```

#### - min

```python
(car_details
    .groupby("fuel")["selling_price"]
    .min()

).astype("uint32")
```

#### - max

```python
(car_details
    .groupby("fuel")["selling_price"]
    .max()

).astype("uint32")
```

#### - count

```python
(car_details
    .groupby("name")["seller_type"]
    .count()

).astype("uint8")
```

#### - unique

```python
(car_details
    .groupby("transmission")["fuel"]

).unique()
```

#### - agg

```python
(car_details
    .groupby("transmission")
    .agg(
        max_km_driven = pd.NamedAgg(column = "km_driven", aggfunc=max),
        min_selling_price = pd.NamedAgg(column = "selling_price", aggfunc=min)
    )
)
```
